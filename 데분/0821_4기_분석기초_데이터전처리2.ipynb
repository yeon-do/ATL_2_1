### groupby, agg 

```python
import pandas as pd
```

```python
exam = pd.read_csv('exam.csv')
```

```python
df_1=exam.groupby('nclass').agg(mean_math=('math','mean'))
```

```python
exam.groupby('nclass', as_index=True).agg(mean_math=('math','mean'))
```

```python
df_1.reset_index() #인덱스를 컬럼으로 바꾸기!
```

### 모든 변수에 대해서 요약 통계

```python
exam.groupby('nclass').mean()
```

```python
exam.groupby('nclass').sum()
```

```python
exam.groupby('nclass')['math','english','science'].sum() #내가 원하는 컬럼을 추가해서 넣을 수 있음
```

```python
mpg = pd.read_csv('mpg.csv')
```

```python
mpg
```

```python
mpg.groupby(['manufacturer','drv']).agg(mean_cty = ('cty','mean'))
```

```python
mpg
```

```python
mpg.query("manufacturer=='audi'").groupby(['drv']).agg(n=('drv','count'))
```

## 데이터 병합 
- merge, concat (join을 이용해서! )

```python
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6]})
df2 = pd.DataFrame({'A':[1,2,3],
                   'C':[7,8,9]})
```

```python
display(df1)
display(df2)
```

```python
#병합 진행
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

- merge

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how='left', on='A')
#on 교집합의 개념 
```

### 교집합이 2개 이상인 경우!

```python
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6],
                   'C':[13,14,15]})
df2 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,7],
                   'D':[10,11,12]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how='inner', on=['A','B']) #정말 교집합만 반환 A, B를 기준으로 모두 다 같아야 한다!
```

```python
df1
df2
```

```python
display(df1)
display(df2)
```

```python
print(df1)
print(df2)
```

### 결측치 찾아보기!

```python
df = pd.DataFrame({'A':[1,2,3,4,5,np.nan],
                  'B':[7,8,np.nan,9,10,11]})
```

### 결측치를 제거하는 경우!

```python
df.dropna() #전체 다 결측치 제거하는 법
```

```python
#내가 원하는 컬럼에 결측치 제거 
df.dropna(subset = ['B']) #B에 대한 결측치만 제거한다.
```

```python
exam.loc[[3,7,12],['science']] = np.nan
```

```python
exam.isna().sum()
```

```python
#결측치를 특정 값으로 대입하는 방법
exam 
#결측치에 대해서 평균 값을 대체한다.
```

```python
exam['science'].mean()
#과학 평균이 61
```

```python
exam['science'] = exam['science'].fillna(61) #fillna na값을 채운다.
```

```python
exam['science'].mean() #na값 대체시
```

```python
exam['science'].mean() 
```

### 평균이 아닌 0으로 채운다

```python
exam['science'] = exam['science'].fillna((4+5)/3) #fillna na값을 채운다. #안에 숫자가 나오는 공식도 사용할 수 있다.
```

```python
exam
```

```python
exam['science'].mean()
```
